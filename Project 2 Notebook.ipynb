## The Dataset: Hospital Inpatient Cost Transparency: Beginning 2009 

These data are provided by New York State Department of Health 

Retrieved from: https://health.data.ny.gov/Health/Hospital-Inpatient-Cost-Transparency-Beginning-200/7dtz-qxmr/data 

Last Updated: Dec. 4, 2019 

This dataset contains information from 28 NYS hospitals on specific charge and cost data for treatments. 

Data for this dataset was submitted by hospitals through the NYS Department of Health system “the statewide planning and Research Cooperative System.” Data elements include the year for the charge, facility information, classification codes for illnesses and severity of illnesses, and whether the charge was for medical, surgical, or other services. Metrics include the total number of discharge and the mean and median charge and cost of services for each case (i.e. each hospital, level of severity, type of illness, etc.).  

## Project Proposal 

Unmet need for mental health services is a huge problem in the U.S., and according to the most recent data from the National Survey of Drug Use and Health (NSDUH), the most common reason for not receiving services for an unmet need is cost (2018 NSDUH, Table 8.33b). As a result, I chose a dataset that would allow me to investigate this primary barrier to care further. 

Using the hospital inpatient cost transparency dataset, I propose an analysis that will look at how treatments specifically for mental health related hospitalizations have changed in cost over time as well as how the number of hospitalizations have changed over time. I would like to compare these values to all other reasons for hospitalization to see if the trends are unique to mental health treatment or consistent across the healthcare system. 

### Hypothesis 1: 
The cost of healthcare overall has been increasing at an unprecedented rate and remains the number one barrier for people accessing needed care. I hypothesize that the cost of mental health care has increased at a greater rate than the cost of other types of health care, particularly in the last few years (2016- present) because of a spike in demand for services associated with the affordable care act and generally increasing rates of mental illness and stress nation wide (2019 APA Stress in America Report; 2018 NSDUH). 

### Hypothesis 2: 
I hypothesize, based on NSDUH data and changes in federal policy (the affordable care act) that the total number of hospitalizations for mental health care have increased at a higher rate than other types of hospitalizations.

The proposed analysis will include a look at aggregate cost and number of discharges for all mental health services for each year, as well as segmented analyses, looking at the costs and numbers for each type of hospitalization.  

Produced figures will include line graphs showing change over time for median cost, and change over time for number of discharges for each of these segments.

Most of the code looks clean and is formated appropriately. There are some missing values though, and I will look to see if these are worth keeping in or dropping. The raw code below investigates these things. I set them to raw to save space after I ran them:

Based on the inability to group these data or make determinations on what the reason for treatment was, I am going to drop those rows.

```python
# Drop the rows with missing data
Clean_DF = dataframe[dataframe['APR Severity of Illness Code'] != 0] #removes all not applicable and nan rows from this variable
Clean_DF = dataframe[dataframe['APR Medical Surgical Code'].isnull() == False]
```

# Analysis and Visualization

Now that my dataset is clean I can start my analysis. First, I will check what variables have an effect on the cost. I will use box plots to compare the median cost by severity of illness, type of illness (mental health or not) and surgical vs. Medical:


```python
# I will start by adding a tag for if it's a mental health reason or not
def new_column(df, column):
    new_list = []
    for i in df[column]:
        if i > 776:
            new_list.append('Not Mental Health')
        elif i < 740:
            new_list.append('Not Mental Health')
        else:
            new_list.append('Mental Health')
    return new_list
        
#Clean_DF['MH Tag'] = new_column(Clean_DF['APR DRG Code'])
Clean_DF.loc[:,"MH Tag"] = new_column(Clean_DF, 'APR DRG Code')
```

```python
#Clean_DF.head().T
```

```python
#First, I will look at mental health versus non mental health costs. I will use the boxplot function from: https://python-graph-gallery.com/39-hidden-data-under-boxplot/
sns.boxplot(x='MH Tag', y='Median Cost', data=Clean_DF)
plt.title('Median Cost of Mental Health vs. Non-Mental Health Treatment in NYS Hospitals')
plt.ylim(0, 50000) # I needed to limit the y axis because there were a significant number of outliers, ranging to almost 8 million in the non-mental health group
```

```python
#Comparing only non-surgical treatment costs
sns.boxplot(x='MH Tag', y='Median Cost', data=(Clean_DF[Clean_DF['APR Medical Surgical Code'] == 'M']))
plt.title('Median Cost of Non-Surgical Treatment for Mental Health vs. Non-Mental Health Treatment in NYS Hospitals')
plt.ylim(0, 50000) # I needed to limit the y axis because there were a significant number of outliers, ranging to almost 8 million in the non-mental health group
```

Conclude that mental health costs are lower and have a smaller range than non-mental health costs overall, but that difference narrows when considering only the non-surgical procedures.

```python
#Next I will look at box plots of the median cost by severity of illness
sns.boxplot(x='APR Severity of Illness Description', y='Median Cost', data=Clean_DF)
plt.title('Median Cost of Treatment in NYS Hospitals By Severity of Illness')
plt.ylim(0, 100000) # I needed to limit the y axis because there were a significant number of outliers
```

Based on the above boxplot, extreme and magor illnesses had higher cost than moderate or minor illnesses. For the purpose of this analysis, I will separate out by severity of illness in order to avoid biases based on the proportion of illnesses that fall into each group (analysis of this is at the end: about 7% of Mental Illness related hospitalizations were Extreme while 17% of non mental illness related hospitalizations were).

```python
#Next I will look at box plots of the median cost by Type of treatment
sns.boxplot(x='APR Medical Surgical Description', y='Median Cost', data=Clean_DF)
plt.title('Median Cost of Treatment in NYS Hospitals By Type of Treatment')
plt.ylim(0, 70000) # I needed to limit the y axis because there were a significant number of outliers
```

The median and range of the median costs for medical procedures was much lower than for surgical. Since mental health treatments are less likely to be surgical in nature, I will proceed with the analysis looking only at medical procedures.

## Hypothesis Testing

<b> Testing my first hypothesis: </b>

My plan is to first, graph the total median cost over all services by year to see if there are any general trends. Then I plan to graph the total median cost for all mental illness related hospitalizations that were medical, against cost for all medical non-mental illness related hospitalization, by year. I will attempt to divide these by severity of illness as well

```python
#remove all non-medical rows from the dataset
Clean_DF = Clean_DF[Clean_DF['APR Medical Surgical Code'] == 'M']
#Clean_DF.head()
```

```python
#Produce graph of average overall medians across all treatment types.
Clean_DF[['Median Cost', 'Year']].groupby(['Year']).mean().plot()
plt.title('Plot of Mean Median Cost of Health Care in NYS Hospitals by Year')
plt.xlabel('Year')
plt.ylabel('Median Cost')
plt.ylim(0.14500)
```

As expected, there is an upward trend in cost which increased in rate around 2012.

```python
#Methods from: https://seaborn.pydata.org/tutorial/categorical.html
#Produce graph comparing results for Mental health vs. Non-mental health treatments
sns.catplot(x="Year", y="Median Cost", hue="MH Tag",
            palette={"Mental Health": "g", "Not Mental Health": "m"},
            markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=Clean_DF[['Median Cost', 'Year', 'MH Tag']].groupby(['Year', 'MH Tag']).mean().reset_index());
plt.ylim(0,20000)
plt.title('Average Median Cost of Medical Treatment in NYS Hospitals by Year and Illness Type')
```

```python
# Next, visualizing the average median cost by severity of illness 
sns.catplot(x="Year", y="Median Cost", hue="APR Severity of Illness Description",
            palette={"Minor": "g", "Moderate": "m", "Major": "y", "Extreme": "b"},
            markers=["^", "o","*", '+'], linestyles=["-", "--","-", "--"],
            kind="point", data=Clean_DF[['Median Cost', 'Year', 'APR Severity of Illness Description']].groupby(['Year', 'APR Severity of Illness Description']).mean().reset_index());
plt.title('Average Median Cost of Medical Treatment in NYS Hospitals by Year and Severity of Illness')
```

For my final figure testing hypothesis 1, I would like to compare two graphs, showing average median cost by severity of illness, for those with mental illness versus those without. 

```python
#Create adjacent graphs by severity of illness for mental health and non-mental health treatments to compare between the two

Clean_DF1 = Clean_DF[Clean_DF['MH Tag'] == 'Mental Health']
Clean_DF2 = Clean_DF[Clean_DF['MH Tag'] == 'Not Mental Health']

sns.catplot(x="Year", y="Median Cost", hue="APR Severity of Illness Description",
            palette={"Minor": "g", "Moderate": "m", "Major": "y", "Extreme": "b"},
            markers=["^", "o","*", '+'], linestyles=["-", "--","-", "--"],
            kind="point", data=Clean_DF1[['Median Cost', 'Year', 'APR Severity of Illness Description']].groupby(['Year', 'APR Severity of Illness Description']).mean().reset_index());
plt.title('Average Median Cost of Mental Health Medical Treatment in NYS Hospitals by Year and Severity of Illness')
plt.ylim(0,40000)

sns.catplot(x="Year", y="Median Cost", hue="APR Severity of Illness Description",
            palette={"Minor": "indigo", "Moderate": "r", "Major": "slategrey", "Extreme": "darkgreen"},
            markers=["^", "o","*", '+'], linestyles=["-", "--","-", "--"],
            kind="point", data=Clean_DF2[['Median Cost', 'Year', 'APR Severity of Illness Description']].groupby(['Year', 'APR Severity of Illness Description']).mean().reset_index());
plt.title('Average Median Cost of Non-Mental Health Medical Treatment in NYS Hospitals by Year and Severity of Illness')
plt.ylim(0,40000)
plt.show()
```

```python
# For the sake of clarity, I wanted to put these two graphs next to eachother to compare directly
# I used guidance from https://seaborn.pydata.org/generated/seaborn.FacetGrid.html#seaborn.FacetGrid to figure out how to do this

g = sns.FacetGrid((Clean_DF[['Median Cost', 'Year', 'APR Severity of Illness Description', 'MH Tag']].groupby(['Year', 'APR Severity of Illness Description', 'MH Tag']).mean().reset_index()), col="MH Tag", col_wrap=2, height=5)
g = g.map(plt.scatter, "Year", "Median Cost", marker=".")
```
