# Load data

* Time: The start time of the event in epoch time format
* Duration: The duration of the event in seconds.
* SrcDevice: The device that likely initiated the event.
* DstDevice: The receiving device.
* Protocol: The protocol number.
* SrcPort: The port used by the SrcDevice.
* DstPort: The port used by the DstDevice.
* SrcPackets: The number of packets the SrcDevice sent during the event.
* DstPackets: The number of packets the DstDevice sent during the event.
* SrcBytes: The number of bytes the SrcDevice sent during the event.
* DstBytes: The number of bytes the DstDevice sent during the event.

# TASK: what number of nodes makes the visualization unreadable?

You have a dataframe with 1,000,000 rows. 

First, design your experiments. What number of rows will you explore to address the question? 

At what point does node and/or edge labeling provide value?

```python
G = nx.from_pandas_edgelist(nfdf[0:100],'SrcDevice','DstDevice','Duration')
```

```python
nx.draw(G, with_labels = False, node_size = 100)
plt.show()
```

```python
pos = nx.spring_layout(G)
nx.draw(G,with_labels=True)
nx.draw_networkx_edge_labels(G,pos=pos,with_labels=True)
_=plt.axis('off')
```

The label version is already unreadable, so I will scale that one back.

```python
G = nx.from_pandas_edgelist(nfdf[0:50],'SrcDevice','DstDevice','Duration')
```

```python
pos = nx.spring_layout(G)
nx.draw(G,with_labels=True)
nx.draw_networkx_edge_labels(G,pos=pos,with_labels=True)
_=plt.axis('off')
```

with 50 rows of data I can now see at least which devices are involved, mostly, which is valuable. I will increase by 25 to see if I lose that clarity

```python
#Overlaid labels make it impossible to read, will halve the difference again to see if it's better
G = nx.from_pandas_edgelist(nfdf[0:62],'SrcDevice','DstDevice','Duration')
pos = nx.spring_layout(G)
nx.draw(G,with_labels=True)
nx.draw_networkx_edge_labels(G,pos=pos,with_labels=True)
_=plt.axis('off')
```

This is still unreadable, therefore, When labels are included, I conclude that around 50 lines of data are as many as can be included to provide value. Next I will ramp up the number of nodes included in the labelless graph, to see at what point interpretation is no longer possible

```python
G = nx.from_pandas_edgelist(nfdf[0:1000],'SrcDevice','DstDevice','Duration')
nx.draw(G, with_labels = False, node_size = 100)
plt.show()
```

at 1000 nodes, I can see clusters, but the relationships between individual nodes are no longer apparent. Despite losing some of the granularity here, there is still information to be gleaned (in the form of clusters and relative cluster sizes), therefore, I will continue to see what happens when I increase the number of rows included

```python
G = nx.from_pandas_edgelist(nfdf[0:10000],'SrcDevice','DstDevice','Duration')
nx.draw(G, with_labels = False, node_size = 100)
plt.show()
```

I can see now that there are several small clusters of nodes, and a few large interconnected clusters as well

```python
G = nx.from_pandas_edgelist(nfdf[0:100000],'SrcDevice','DstDevice','Duration')
nx.draw(G, with_labels = False, node_size = 100)
plt.show()
```

I've definitely reached the end of being able to extract more meaningful information or interpretation from these data, as the patterns are staying the same. In terms of big picture, I would say that this visual is the most valuable around 10,000 rows, because you can still see some of the network structure, without it being obscured completely by the nodes.
